## Ключевые функции (текущая версия)

1. **Загрузка плана** — изображение, PDF, DWG (конвертация в PNG)
2. **Точка отсчёта и направление** — клик по плану, коррекция координат при object-contain
3. **Загрузка видео** — стриминг, таймаут 30 мин
4. **Анализ по video_id** — двухшаговый процесс: загрузка → анализ
5. **Библиотека видео** — загрузка сохранённого анализа
6. **Pathfinding** — траектория следует плану, не проходит сквозь стены
7. **Компас** — угол поворота плана, калибровка направления
8. **Редактор плана** — рисование стен и прямоугольников
9. **Библиотека планов** — сохранённые планы
